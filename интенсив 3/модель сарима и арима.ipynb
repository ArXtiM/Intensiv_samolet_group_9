```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt
from pmdarima import auto_arima
import warnings
warnings.filterwarnings('ignore')

# Загрузка данных
data = pd.read_csv('cleaned_data.csv', parse_dates=['Date'], index_col='Date')
ts = data['Цена на арматуру']

# Визуализация временного ряда с разбивкой по годам
plt.figure(figsize=(14,7))
plt.plot(ts)
plt.title('Цена на арматуру с 2017 по 2022 год', fontsize=14)
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Цена', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
years = pd.date_range(start=ts.index.min(), end=ts.index.max(), freq='YS')
for year in years:
    plt.axvline(year, color='gray', linestyle='--', alpha=0.5)
plt.show()

# Тест Дики-Фуллера с более подробным выводом
def test_stationarity(timeseries):
    print('Результаты теста Дики-Фуллера:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Тестовая статистика','p-value','Количество лагов','Количество наблюдений'])
    for key,value in dftest[4].items():
        dfoutput[f'Критическое значение ({key})'] = value
    print(dfoutput)
    
    if dfoutput['p-value'] > 0.05:
        print("\nРяд нестационарен. Рекомендуется дифференцирование.")
    else:
        print("\nРяд стационарен. Дифференцирование не требуется.")

test_stationarity(ts)

# Автоматический подбор параметров ARIMA с помощью pmdarima
print("\nАвтоматический подбор параметров ARIMA...")
model_auto = auto_arima(ts, start_p=1, start_q=1,
                        max_p=5, max_q=5, m=12,
                        start_P=0, seasonal=True,
                        d=None, D=None, trace=True,
                        error_action='ignore',  
                        suppress_warnings=True, 
                        stepwise=True)

print(model_auto.summary())

# Разделение на обучающую и тестовую выборки
train_size = int(len(ts) * 0.8)
train, test = ts[0:train_size], ts[train_size:]

# Создание и обучение лучшей ARIMA модели на основе auto_arima
best_order = model_auto.order
best_seasonal_order = model_auto.seasonal_order

print(f"\nЛучшие параметры ARIMA: order={best_order}, seasonal_order={best_seasonal_order}")

# Модель SARIMA с лучшими параметрами
model_sarima = SARIMAX(train, 
                      order=best_order,
                      seasonal_order=best_seasonal_order,
                      enforce_stationarity=False,
                      enforce_invertibility=False)

model_sarima_fit = model_sarima.fit(disp=False)
print(model_sarima_fit.summary())

# Анализ остатков модели
residuals = model_sarima_fit.resid
plt.figure(figsize=(12,6))
plt.plot(residuals)
plt.title('Остатки модели SARIMA', fontsize=14)
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Ошибка', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# График распределения остатков
plt.figure(figsize=(12,6))
plt.hist(residuals, bins=30, edgecolor='k', alpha=0.65)
plt.title('Распределение остатков модели SARIMA', fontsize=14)
plt.xlabel('Ошибка', fontsize=12)
plt.ylabel('Частота', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Прогнозирование
sarima_forecast = model_sarima_fit.get_forecast(steps=len(test))
sarima_forecast_conf = sarima_forecast.conf_int()
sarima_forecast_values = sarima_forecast.predicted_mean

# Оценка качества
rmse = sqrt(mean_squared_error(test, sarima_forecast_values))
r2 = r2_score(test, sarima_forecast_values)
mape = np.mean(np.abs((test - sarima_forecast_values) / test)) * 100

print(f"\nМетрики качества SARIMA модели:")
print(f"RMSE: {rmse:.2f}")
print(f"R²: {r2:.4f}")
print(f"MAPE: {mape:.2f}%")

# Визуализация прогноза с доверительным интервалом
plt.figure(figsize=(14,7))
plt.plot(train, label='Обучающая выборка', color='blue')
plt.plot(test, label='Фактические значения', color='green')
plt.plot(sarima_forecast_values, label='Прогноз SARIMA', color='red')
plt.fill_between(sarima_forecast_conf.index,
                sarima_forecast_conf.iloc[:, 0],
                sarima_forecast_conf.iloc[:, 1], 
                color='pink', alpha=0.3, label='Доверительный интервал 95%')

plt.title('Прогнозирование цены на арматуру с помощью SARIMA', fontsize=14)
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Цена', fontsize=12)
plt.legend(loc='upper left', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Сравнение с простой ARIMA моделью (без сезонности)
print("\nСравнение с простой ARIMA моделью...")
model_arima = ARIMA(train, order=(2,1,2))
model_arima_fit = model_arima.fit()
arima_forecast = model_arima_fit.forecast(steps=len(test))

# Оценка качества ARIMA
rmse_arima = sqrt(mean_squared_error(test, arima_forecast))
r2_arima = r2_score(test, arima_forecast)
mape_arima = np.mean(np.abs((test - arima_forecast) / test)) * 100

print(f"\nМетрики качества ARIMA модели:")
print(f"RMSE: {rmse_arima:.2f}")
print(f"R²: {r2_arima:.4f}")
print(f"MAPE: {mape_arima:.2f}%")

# Итоговое сравнение моделей
comparison = pd.DataFrame({
    'Метрика': ['RMSE', 'R²', 'MAPE (%)'],
    'SARIMA': [f"{rmse:.2f}", f"{r2:.4f}", f"{mape:.2f}"],
    'ARIMA': [f"{rmse_arima:.2f}", f"{r2_arima:.4f}", f"{mape_arima:.2f}"]
})

print("\nСравнение моделей:")
print(comparison)

# Визуализация сравнения моделей
plt.figure(figsize=(14,7))
plt.plot(test, label='Фактические значения', color='black', linewidth=2)
plt.plot(sarima_forecast_values, label=f'SARIMA (R²={r2:.3f})', color='red')
plt.plot(arima_forecast, label=f'ARIMA (R²={r2_arima:.3f})', color='blue')
plt.title('Сравнение моделей ARIMA и SARIMA', fontsize=14)
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Цена', fontsize=12)
plt.legend(loc='upper left', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller

# Загрузка данных
data = pd.read_csv('C:\\Users\\VikA\\Desktop\\интенсив 3\\merged_data.csv', sep=';', parse_dates=['Date'], index_col='Date')

# Выбор временного ряда для анализа (например, CHMF_Price)
ts = data['CHMF_Price'].dropna()

# Проверка на стационарность
def check_stationarity(timeseries):
    result = adfuller(timeseries)
    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))

check_stationarity(ts)

# Если ряд не стационарный, применяем дифференцирование
ts_diff = ts.diff().dropna()
check_stationarity(ts_diff)

# Построение модели ARIMA
# Подбор параметров (p, d, q) может быть выполнен с использованием ACF и PACF графиков
plot_acf(ts_diff)
plot_pacf(ts_diff)
plt.show()

# Пример параметров (p, d, q) для ARIMA
model_arima = ARIMA(ts, order=(1, 1, 1))
results_arima = model_arima.fit()
print(results_arima.summary())

# Построение модели SARIMA
# Пример параметров (p, d, q) и сезонных параметров (P, D, Q, S) для SARIMA
model_sarima = SARIMAX(ts, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
results_sarima = model_sarima.fit()
print(results_sarima.summary())

# Прогнозирование на будущие периоды
forecast_arima = results_arima.forecast(steps=10)
forecast_sarima = results_sarima.forecast(steps=10)

print("ARIMA Forecast:", forecast_arima)
print("SARIMA Forecast:", forecast_sarima)

# Визуализация результатов
plt.figure(figsize=(12, 6))
plt.plot(ts, label='Original')
plt.plot(forecast_arima, label='ARIMA Forecast')
plt.plot(forecast_sarima, label='SARIMA Forecast')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
from sklearn.metrics import mean_squared_error
from math import sqrt

# Загрузка данных
data = pd.read_csv('cleaned_data.csv', parse_dates=['Date'], index_col='Date')
ts = data['Цена на арматуру']

# Визуализация временного ряда
plt.figure(figsize=(12,6))
plt.plot(ts)
plt.title('Цена на арматуру с 2017 по 2022 год')
plt.xlabel('Дата')
plt.ylabel('Цена')
plt.grid()
plt.show()

# Тест Дики-Фуллера
result = adfuller(ts)
print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:')
for key, value in result[4].items():
    print(f'\t{key}: {value}')

# Если p-value > 0.05, ряд нестационарен
if result[1] > 0.05:
    print("Ряд нестационарен, требуется дифференцирование")
else:
    print("Ряд стационарен")


# Графики ACF и PACF
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12,8))
plot_acf(ts, lags=40, ax=ax1)
plot_pacf(ts, lags=40, ax=ax2)
plt.show()

# Разделение на обучающую и тестовую выборки
train_size = int(len(ts) * 0.8)
train, test = ts[0:train_size], ts[train_size:]

# Подбор параметров ARIMA (p,d,q)
# На основе ACF/PACF и теста Дики-Фуллера выбираем d=1
model_arima = ARIMA(train, order=(2,1,2))
model_arima_fit = model_arima.fit()
print(model_arima_fit.summary())

# Прогнозирование
arima_forecast = model_arima_fit.forecast(steps=len(test))

# Оценка качества
rmse = sqrt(mean_squared_error(test, arima_forecast))
print(f'RMSE ARIMA: {rmse}')

# Визуализация прогноза
plt.figure(figsize=(12,6))
plt.plot(train, label='Обучающая выборка')
plt.plot(test, label='Тестовая выборка')
plt.plot(arima_forecast, label='Прогноз ARIMA')
plt.legend()
plt.title('Прогнозирование цены на арматуру с помощью ARIMA')
plt.show()

# SARIMA учитывает сезонность (если она есть)
# Параметры: (p,d,q)(P,D,Q,s)
# s - длина сезонного цикла (например, 52 для недельных данных за год)

# Попробуем несколько вариантов сезонности
model_sarima = SARIMAX(train, order=(1,1,1), seasonal_order=(1,1,1,52))
model_sarima_fit = model_sarima.fit()
print(model_sarima_fit.summary())

# Прогнозирование
sarima_forecast = model_sarima_fit.forecast(steps=len(test))

# Оценка качества
rmse = sqrt(mean_squared_error(test, sarima_forecast))
print(f'RMSE SARIMA: {rmse}')

# Визуализация прогноза
plt.figure(figsize=(12,6))
plt.plot(train, label='Обучающая выборка')
plt.plot(test, label='Тестовая выборка')
plt.plot(sarima_forecast, label='Прогноз SARIMA')
plt.legend()
plt.title('Прогнозирование цены на арматуру с помощью SARIMA')
plt.show()



plt.figure(figsize=(12,6))
plt.plot(test, label='Фактические значения')
plt.plot(arima_forecast, label='Прогноз ARIMA')
plt.plot(sarima_forecast, label='Прогноз SARIMA')
plt.legend()
plt.title('Сравнение моделей ARIMA и SARIMA')
plt.show()
```
